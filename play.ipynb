```python
query_option_search = "2 inch, feed animals"
search = None
if query_option_search:
    phrases = query_option_search.split(",")
    for phrase in phrases:
        # Remove leading and trailing whitespaces from the phrase
        phrase = phrase.strip()
        search = f"{search} \"{phrase}\"" if search else f"\"{phrase}\"" 
print(search)
```
